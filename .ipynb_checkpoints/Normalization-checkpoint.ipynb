```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
df = pd.read_csv('SBAnational.csv')
df
```

```python
type(df['DisbursementGross'][0])
```

```python
df.info()
```

```python
box = list(df.columns)
out_df = pd.DataFrame()
for feature in box:
    if (type(df[feature][0]) == np.int64):
        data = pd.DataFrame(df[feature])
        scaler = MinMaxScaler()
        normalized = scaler.fit_transform(data)
        out_df[feature] = [float(x) for x in normalized]
out_df
```
