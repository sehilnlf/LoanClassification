```python
box = [
 'Term',
 'NoEmp',
 'GrAppv',
 'SBA_Appv']
out_df = df.copy()
for feature in box:
        data = pd.DataFrame(df[feature])
        scaler = MinMaxScaler(feature_range=(0, 1))
        normalized = scaler.fit_transform(data)
        out_df[feature] = [float(x) for x in normalized]
out_df
```

```python
out_df.to_csv('data_normalized.csv')
```
